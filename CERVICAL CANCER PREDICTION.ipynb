```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset =pd.read_csv("cervical-cancer1.csv")
x=dataset.iloc[:, :-1].values
y=dataset.iloc[:,-1].values
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split

# Assuming 'x' is a pandas DataFrame and 'y' is the target variable
df = pd.DataFrame(x)  # Convert 'x' to DataFrame if it's not already

# Replace 'NaN' string or any other placeholder with actual NaN values
df = df.replace('NaN', pd.NA)

# Create SimpleImputer with 'mean' strategy
imputer = SimpleImputer(strategy='mean')

# Apply imputer to the DataFrame
df_imputed = imputer.fit_transform(df)  # This applies to all columns by default

# Convert back to DataFrame with original column names
df_imputed = pd.DataFrame(df_imputed, columns=df.columns)

# Now, split the imputed data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(df_imputed, y, test_size=0.30, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)
```

```python
def models(x_train,y_train):
  
  #Using Logistic Regression Algorithm to the Training Set
  from sklearn.linear_model import LogisticRegression
  log = LogisticRegression(random_state = 12345)
  log.fit(x_train, y_train)

  #Using KNeighborsClassifier Method of neighbors class to use Nearest Neighbor algorithm
  from sklearn.neighbors import KNeighborsClassifier
  knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
  knn.fit(x_train, y_train)

  #Using SVC method of svm class to use Support Vector Machine Algorithm
  from sklearn.svm import SVC
  svc_lin = SVC(kernel = 'linear', random_state = 0)
  svc_lin.fit(x_train, y_train)

  #Using SVC method of svm class to use Kernel SVM Algorithm
  from sklearn.svm import SVC
  svc_rbf = SVC(kernel = 'rbf', random_state = 0)
  svc_rbf.fit(x_train, y_train)

  #Using GaussianNB method of naÃ¯ve_bayes class to use NaÃ¯ve Bayes Algorithm
  from sklearn.naive_bayes import GaussianNB
  gauss = GaussianNB()
  gauss.fit(x_train, y_train)

  #Using DecisionTreeClassifier of tree class to use Decision Tree Algorithm
  from sklearn.tree import DecisionTreeClassifier
  tree = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
  tree.fit(x_train, y_train)

  #Using RandomForestClassifier method of ensemble class to use Random Forest Classification algorithm
  from sklearn.ensemble import RandomForestClassifier
  forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
  forest.fit(x_train, y_train)
  
  #print model accuracy on the training data.
  print('[0]Logistic Regression Training Accuracy:', log.score(x_train, y_train))
  print('[1]K Nearest Neighbor Training Accuracy:', knn.score(x_train, y_train))
  print('[2]Support Vector Machine (Linear Classifier) Training Accuracy:', svc_lin.score(x_train, y_train))
  print('[3]Support Vector Machine (RBF Classifier) Training Accuracy:', svc_rbf.score(x_train, y_train))
  print('[4]Gaussian Naive Bayes Training Accuracy:', gauss.score(x_train, y_train))
  print('[5]Decision Tree Classifier Training Accuracy:', tree.score(x_train, y_train))
  print('[6]Random Forest Classifier Training Accuracy:', forest.score(x_train, y_train))
  
  return log, knn, svc_lin, svc_rbf, gauss, tree, forest
```

```python
model = models(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

# Assuming 'model' is a list of trained models
for i in range(len(model)):
    # Predict the labels on the test data using the current model
    y_pred = model[i].predict(x_test)

    # Calculate the confusion matrix
    cm = confusion_matrix(y_test, y_pred)

    # Calculate the accuracy of the model
    accuracy = accuracy_score(y_test, y_pred)

    # Print confusion matrix and accuracy for the model
    print(f"Model {i+1} - Confusion Matrix:")
    print(cm)
    print(f"Model {i+1} - Accuracy: {accuracy:.2f}\n")
```
